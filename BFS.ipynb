```python
def bfs(graph, start, target):
    """
    This function returns the path from the start node to the goal node in a given connected graph.
    It searches level by level for the goal node.

    Parameter graph: Graph conataining the start node and the goal node 
    Precondition: Type is Graph object and must be connected
    
    Parameter start: it is the start node
    Precondition: Type is string and must be in the nodes of the graph 

    Parameter Target: The goal node
    Precondition: Type is string and must be in the nodes of the graph 
    """
    if start not in graph or target not in graph:
        return None  # One of the nodes doesn't exist in the graph

    visited = []
    queue = [[start]]

    while queue:
        path = queue.pop(0)
        node = path[-1]

        if node == target:
            return path  # if node Found the target node, return the path

        if node not in visited:
            visited.append(node)
            for neighbor in graph.get(node, []):
                new_path = path.copy()  # Create a new path
                new_path.append(neighbor)
                queue.append(new_path)

    return None  # Target node is not reachable from the start node

# Example for the above code

start_node = "Lucknow"
target_node = "Mirzapur"

path_to_target = bfs(graph, start_node, target_node)

print(path_to_target)
```
