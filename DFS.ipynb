```python
def DFS(graph, start, target, path=None):
    """
    This function returns the path from the start node to the goal node in a given connected graph.
    It searches Branch by branch. Goes to the depth of one branch andif didn't fiond the goal, backtrackes.

    Parameter graph: Graph conataining the start node and the goal node 
    Precondition: Type is Graph object and must be connected
    
    Parameter start: it is the start node
    Precondition: Type is string and must be in the nodes of the graph 

    Parameter Target: The goal node
    Precondition: Type is string and must be in the nodes of the graph 
    """
    if path is None:
        path = []
    
    path.append(start)
    if start == target:
        return path

    for neighbor in graph.get(start, []):
        if neighbor not in path:
            new_path = DFS(graph, neighbor, target, path.copy()) # Recusrion on a branch
            if new_path:
                return new_path
    return None


# Example for the above code

start_node = "Lucknow"
target_node = "Mirzapur"

path = DFS(graph, start_node, target_node)

if path:
    print(f"Path from Node {start_node} to Node {target_node}: {path}")
else:
    print(f"No path found from Node {start_node} to Node {target_node}.")
```
